```python
import math
def calcularfx(x):
    fx = (1 / (x ** 10)) * (math.sin(x) + (5 * math.exp(2 * x))) ** (1 / 4)
    return fx

def volumen(di, al):
    v = math.pi * (di / 2) ** 2 * al
    return v

def area(bme, bma, hip):
    h = (hip ** 2 - ((bma - bme) / 2) ** 2) ** (1 / 2)
    a = ((bma + bme) / 2) * h
    p = bma + bme + hip
    return a,p

while True:
    print('A.- calcular funcion f de x')
    print('B.- Volumen del cilindro')
    print('C.- area y perimetro del trapecio regular')
    print('D.- Salir')
    p = input()

    if p == 'A':
        x = float(input('Ingrese valor de x: '))
        print('El valor de Fx es', calcularfx(x))
    elif p == 'B':
        di = float(input('Ingrese valor del diametro: '))
        al = float(input('Ingrese valor de la altura: '))
        vol = volumen(di, al)
        print('El volumen es', vol)
    elif p == 'C':
        bme = float(input('Ingrese valor de Base menor: '))
        bma = float(input('Ingrese valor de Base mayor: '))
        hip = float(input('Ingrese valor de la hipotenusa: '))
        a, p = area(bme, bma, hip)
        print('El area es', a, 'y el perimetro es', p)
    elif p == 'D':
        break
```
