# Enunciado 1

Una Universidad esta interesada en implementar un sistema de enseñanza 
a distancia. El estudio consiste en encuestar a los estudiantes a medida 
que vayan saliendo del comedor, solicitándoles que respondan a las 
siguientes preguntas: ¿nombre?, ¿edad?, ¿Carrera que estudia? Y 
además:
¿Como preferiría ver una materia?: en la WEB__Presencial__ Semipresencial __.
¿Tiene Computador? Si______ No______.
¿Dispone de Conexión a Internet? Si______ No______.
Usando los datos suministrados por los encuestados y considerando que al 
menos un estudiante respondió a las preguntas, diseñe un algoritmo que permita 
obtener la siguiente información:
1. Total de estudiantes que prefirieron la modalidad presencial.
2. Cantidad de estudiantes de Ing. Mecánica que desearían ver una 
materia en la WEB, dado que disponen de computador i conexión a Internet.
3. Promedio de edad de los estudiantes que prefieren tomar una 
materia e forma semipresencial.
4. Nombre del estudiante más joven de Informática, que desea ver la 
materia por la WEB, pero no dispone de computador ni conexión a Internet. 
Considere que existe solo uno.

## Codigo **Python**

```python
from ast import Continue
menor = 200
contprese = 0
acuedad = 0
contmec = 0
contsemipre = 0

while True:
  print("Ingrese el nombre")
  nombre = input()
  print("Ingrese la edad")
  edad = int(input())
  print("Ingrese la carrera")
  carrera = input()
  print("Como Preferiría tomar la Materia 1 web, 2 presencial, 3 Semipresencial")
  while True:
     materia = int(input())
     if materia>1 and materia<4:
       break;   
  while True:
    print("Tiene Computador S/N")
    comp = input()
    if comp not in ['S','s','N','n']:
       print("solo acepto S, N o s, n como respuesta")
    else:
      break
  while True:
   internet = input("Conexión a Internet S/N ")
   if internet in ["S", "N", "s", "n"]:
     break;     
  if materia == 2:
    contprese = contprese + 1
  if carrera == "ing Mecánica" and materia == 1 and comp == 'S' and internet == 'S':
    contmec = contmec + 1
  if materia == 3:
    contsemipre = contsemipre + 1
    acuedad = acuedad + edad
  if carrera == "Informatica" and materia == 1 and comp == 'N' and internet == 'N':
    if edad < menor:
        menor = edad
  while True:
     resp = input("Desea Ingresar Otra encuesta S/N: ")
     if resp in ["S", "N", "s", "n"]:
        break; 
  if resp == "N" or resp == "n":
     break
print("Estudiantes modalidad presencial", contprese)
print("Estudiantes de mecánica ", contmec)
print("Promedio de edad semipresencial", acuedad/contsemipre)
print("Estudiantes mas joven Informatica", menor)
```
